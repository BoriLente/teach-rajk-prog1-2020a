```python
def megoldás(breaks):
    minimum=0
    maximum=100
    tojások_száma=2
    a=0
    n=0
    c=0
    lépésközök=[]
    
    while a<maximum-minimum:
        n+=1
        lépésközök.append(n)
        a+=n
    lépésközök.reverse()
    
    while True:
        minimum+=lépésközök[c]
        c+=1
        if breaks(minimum):
            minimum-=lépésközök[c-1]
            while True:
                minimum+=1
                if breaks(minimum):
                    return minimum-1
                    break
```

```python
eggdrop_100floor_2egg.evaluate(megoldás)
```
